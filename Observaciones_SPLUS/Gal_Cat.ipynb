```python
import subprocess
import os
```

```python
def match_gaia(group):
    ifmt1 = 'csv'
    ifmt2 = 'csv'
    out = f'/home/seba/Documents/CG/Observaciones_SPLUS/Galaxies_{group}'
    ofmt = 'csv'
    ra1 = 'RA'
    dec1 = 'DEC'
    ra2 = '_RAJ2000'
    dec2 = '_DEJ2000'
    error = 5
    in1 = f'/home/seba/Documents/CG/Observaciones_SPLUS/Catalogos_raw/{group}'
    in2 = '/home/seba/Documents/CG/Hernandez-Fernandez/Galaxy_catalogue.csv'
    subprocess.run(f'stilts tskymatch2 ifmt1={ifmt1} ifmt2={ifmt2} out={out} ofmt={ofmt} ra1={ra1} dec1={dec1} ra2={ra2} dec2={dec2} error={error} in1={in1} in2={in2}', shell=True)
```

```python
archivos = os.listdir('/home/seba/Documents/CG/Observaciones_SPLUS/Catalogos_raw')
```

```python
for group in archivos:
    match_gaia(group)
```

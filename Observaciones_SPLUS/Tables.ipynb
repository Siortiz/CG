```python
tab = Table.read('/home/seba/Documents/CG/Observaciones_SPLUS/galaxies_proposal.csv', format='ascii')
tab = Table([tab['Group'], tab['RAJ2000'], tab['DEJ2000']])
filas_seleccionadas = []

# Obtener una lista nica de valores en la columna 'group'
grupos_unicos = set(tab['Group'])

# Iterar sobre cada grupo nico
for grupo in grupos_unicos:
    # Obtener las filas que corresponden al grupo actual
    filas_grupo_actual = tab[tab['Group'] == grupo]
    
    # Seleccionar las dos primeras filas del grupo actual
    filas_seleccionadas.extend(filas_grupo_actual[:2])

# Crear una nueva tabla con las filas seleccionadas
tabla_seleccionada = Table(rows=filas_seleccionadas, names=tab.colnames)

# Imprimir la nueva tabla
print(tabla_seleccionada)
tabla_seleccionada.write('prop_targets.csv', format='csv')
```
